```python
#add values usinng column name
df['Total']=df['HP']+df['Attack']+df['Defense']
df.head(2)

#Drop a column
#df=df.drop(columns='Total')
#df.head(2)

#Add horizontally using indexes
#df['Total']=df.iloc[:,4:10].sum(axis=1)
#df.head(3)

# Adding a row
df._append({'Name':'Pingu'}, ignore_index=True)

#Drop a row
# df.drop(index=)

#sort values
df.sort_values('Name',ascending=False)
cols=list(df.columns.values)
df=df[cols[0:4]+[cols[-1]]+cols[4:12]]
df
```

Filtering Data using loc and iloc

```python
df.loc[8]
```

```python
#loc uses labels to search data
print(df.loc[4,'Attack'])

print(df.loc[4])

print(df.loc[:,'Type 1'])
#Filter by creating highpower variable
Highpower=(df['Attack'] & df['Sp. Atk'])>50

df[Highpower]
# Filter to get specific columns with highpower filter
df.loc[Highpower,['Name','Type 1','Type 2']]

#Different Example
filt=df['Name'].str.contains('saur',na=False)
df.loc[filt,'Name']
#filter based on names strating with 'pi' using rregex functions
dilt= df['Name'].str.contains('^pi[a-z]',flags=re.I,regex=True)
df[dilt]

#Conditional changes or setting value(here we change all the fire entries to flame in type 1)
df.loc[df['Type 1']=='Fire','Type 1']='Flame'
df.loc[df['Total']>300,['Generation','Legendary']]=['New','Mythic']
df.loc[df['Total']>300]
```
